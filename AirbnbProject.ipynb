```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import defaultdict
from nltk.corpus import stopwords
import string
import matplotlib.pyplot as plt
import seaborn as sns

# Download NLTK
nltk.download('punkt')
nltk.download('vader_lexicon')
nltk.download('stopwords')

# Sentiment Analyzer
sid = SentimentIntensityAnalyzer()

# Define keywords/phrases for each feature in feature dictionary
feature_keywords = {
    'good_location': [
        'good location', 'great location', 'perfect location', 'convenient location', 'prime location',
        'central location', 'excellent location', 'ideal location', 'close to everything', 'convenient spot',
        'close to attractions', 'accessible location'],
    'good_host': [
        'good host', 'great host', 'friendly host', 'helpful host', 'amazing host', 'wonderful host',
        'accommodating host', 'welcoming host', 'attentive host', 'kind host', 'responsive host', 'professional host'],
    'good_amenities': [
        'good amenities', 'great amenities', 'nice amenities', 'well-equipped', 'top-notch amenities',
        'excellent amenities', 'superb amenities', 'modern amenities', 'high-quality amenities', 'luxurious amenities',
        'comprehensive amenities', 'everything you need'],
    'good_price': [
        'good price', 'great price', 'affordable', 'cheap', 'worth the price', 'reasonable price',
        'competitive price', 'fair price', 'value for money', 'great value', 'economical', 'budget-friendly', 'good deal'],
    'good_cleanliness': [
        'clean', 'very clean', 'spotless', 'neat', 'well maintained', 'immaculate', 'pristine', 'hygienic',
        'tidy', 'clean and tidy', 'very neat', 'super clean', 'meticulously clean']}

stop_words = set(stopwords.words('english'))

# Preprocess text and filter out stop words
def preprocess_text(text):
    if not isinstance(text, str):
        return []
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return tokens

# Function to check for negative words
def contains_negation(text, keyword):
    negations = ['not', 'never']
    words = text.split()
    for i, word in enumerate(words):
        if word in negations and i + 1 < len(words) and words[i + 1] in keyword:
            return True
    return False

def process_chunk(chunk):
    chunk['tokens'] = chunk['comments'].apply(preprocess_text)
    chunk['sentiment'] = chunk['comments'].apply(lambda x: sid.polarity_scores(x)['compound'] if isinstance(x, str) else 0)

    # Identify positive and negative reviews
    positive_reviews = chunk[chunk['sentiment'] > 0.5]
    negative_reviews = chunk[chunk['sentiment'] < -0.5]

    # Filter out negative keywords in positive reviews
    non_negated_positive_reviews = positive_reviews[
        positive_reviews['comments'].apply(lambda x: not any(contains_negation(x, keyword) for keywords in feature_keywords.values() for keyword in keywords))  ]

    results = defaultdict(lambda: defaultdict(int))
    feature_counts = defaultdict(lambda: defaultdict(int))

    total_reviews_counts = defaultdict(int)
    positive_sentiment_counts = defaultdict(int)
    negative_sentiment_counts = defaultdict(int)

    for listing_id, group in non_negated_positive_reviews.groupby('listing_id'):
        total_reviews_counts[listing_id] += len(group)
        positive_sentiment_counts[listing_id] += len(group)

        for _, row in group.iterrows():
            comments = row['comments']
            if isinstance(comments, str):
                for feature, keywords in feature_keywords.items():
                    for keyword in keywords:
                        if keyword in comments.lower():
                            feature_counts[listing_id][feature] += 1

    for listing_id, group in negative_reviews.groupby('listing_id'):
        total_reviews_counts[listing_id] += len(group)
        negative_sentiment_counts[listing_id] += len(group)

        for _, row in group.iterrows():
            comments = row['comments']
            if isinstance(comments, str):
                for feature, keywords in feature_keywords.items():
                    for keyword in keywords:
                        if contains_negation(comments.lower(), keyword):
                            feature_counts[listing_id][f"not_{feature}"] += 1

    for listing_id in total_reviews_counts:
        results[listing_id]['positive_sentiment'] = positive_sentiment_counts[listing_id]
        results[listing_id]['negative_sentiment'] = negative_sentiment_counts[listing_id]
        results[listing_id]['total_reviews'] = total_reviews_counts[listing_id]

    return results, feature_counts

# consistent color palette for features
feature_colors = {
    'good_location': 'blue',
    'good_host': 'green',
    'good_amenities': 'red',
    'good_price': 'purple',
    'good_cleanliness': 'orange',
    'not_good_location': 'lightblue',
    'not_good_host': 'lightgreen',
    'not_good_amenities': 'lightcoral',
    'not_good_price': 'violet',
    'not_good_cleanliness': 'darkgreen'}

# data files and their corresponding cities
data_files = {
    'nyc': '/content/drive/MyDrive/nycreviews.csv',
    'amsterdam': '/content/drive/MyDrive/amsterdamreviews.csv',
    'london': '/content/drive/MyDrive/londonreviews.csv',
    'singapore': '/content/drive/MyDrive/singaporereviews.csv',
    'sandiego': '/content/drive/MyDrive/sandiegoreviews.csv',
    'tokyo': '/content/drive/MyDrive/tokyoreviews.csv'}

chunk_size = 1000

# Process each city's reviews separately and generate plots
for city, data in data_files.items():
    city_results = defaultdict(lambda: defaultdict(int))
    city_feature_counts = defaultdict(lambda: defaultdict(int))

    for chunk in pd.read_csv(data, chunksize=chunk_size):
        results, feature_counts = process_chunk(chunk)
        for listing_id, data in results.items():
            for key, value in data.items():
                city_results[listing_id][key] += value
        for listing_id, data in feature_counts.items():
            for key, value in data.items():
                city_feature_counts[listing_id][key] += value

    #Filter listings with negative sentiment and "not good" features
    negative_feature_listings = {listing_id: data for listing_id, data in city_feature_counts.items() if any(f"not_{feature}" in data for feature in feature_keywords.keys())}
    top_5_negative_listings = sorted(negative_feature_listings.keys(), key=lambda x: city_results[x]['total_reviews'], reverse=True)[:5]

    print(f"Results for the top 5 most reviewed listings with sentiment features in {city}:")
    for listing_id in top_5_negative_listings:
        print(f"Listing {listing_id}:")
        print(f"  Total Reviews: {city_results[listing_id]['total_reviews']}")
        print(f"  Positive Reviews: {city_results[listing_id]['positive_sentiment']}")
        print(f"  Negative Reviews: {city_results[listing_id]['negative_sentiment']}")
        print("  Feature Mentions:")
        for feature, count in city_feature_counts[listing_id].items():
            if feature.startswith("not_"):
                print(f"    {feature}: {count}")
        print()

    # Plot Feature Mentions for Top 5 Listings with Sentiments for this each city
    plt.figure(figsize=(12, 8))
    feature_counts_melted = pd.DataFrame.from_dict(city_feature_counts, orient='index').loc[top_5_negative_listings].reset_index().melt(id_vars='index', var_name='Feature', value_name='Count')

    sns.barplot(
        x='index',
        y='Count',
        hue='Feature',
        data=feature_counts_melted,
        palette=feature_colors,  # Use the consistent color palette
        hue_order=feature_colors.keys()    )
    plt.xticks(rotation=90)
    plt.title(f'Feature Mentions for Top 5 Most Reviewed Listings in {city.capitalize()}')
    plt.xlabel('Listing ID')
    plt.ylabel('Count')
    plt.tight_layout()

    plt.show()
```
